```python
import numpy as np
import matplotlib.pyplot as plt

fig, (ax, ax2) = plt.subplots(1, 2, figsize=(10, 5), layout='constrained')

x = np.linspace(0,1,1000)
y = x + 0.2*np.sin(2*np.pi*x)
ax.plot(x,y)
ax.plot(x,x, c='black', linestyle='--', linewidth=0.8)
ax.set(xlabel='$x$ (more volatile)', ylabel='y', title='positive azeotrope')

y = x - 0.15*np.sin(2*np.pi*x)
ax2.plot(x,y)
ax2.plot(x,x, c='black', linestyle='--', linewidth=0.8)
ax2.set(xlabel='$x$ (more volatile)', ylabel='y', title='negative azeotrope')

plt.savefig('xy_azeotropes.pdf')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import R

A = 2.5 
B = 0.5 
alpha1 = A + 3*B
alpha2 = A - 3*B
beta1 = -4*B
beta2 = 4*B

x1 = np.linspace(0.001, 0.99999,1000)
x2 = 1 - x1
gamma1 = ((alpha1 * (x2**2) + beta1 * (x2**3)))
gamma2 = ((alpha2 * (x1**2) + beta2 * (x1**3)))

fig,ax = plt.subplots(dpi=300)

plt.plot(x1, x1*np.exp(gamma1))
plt.plot(x1, x2*np.exp(gamma2))
plt.ylim(0,1.4)
plt.xlim(0,1)
plt.legend(['component 1', 'component 2'])
```

```python
from scipy.optimize import fsolve

y1 = x1 * np.exp(gamma1)
y2 = x2 * np.exp(gamma2)

def find_intersections(y1, y2, x1):
    """Finds the compositions where y1 = y2 indicating phase equilibrium."""
    diff = y1 - y2
    zero_crossings = np.where(np.diff(np.sign(diff)))[0]
    return x1[zero_crossings]


equilibrium_x1 = find_intersections(y1, y2, x1)

fig, ax = plt.subplots(dpi=300)
plt.plot(x1, y1, label='component 1')
plt.plot(x1, y2, label='component 2')
plt.axvline(equilibrium_x1[0], linestyle='--', color='red', label=f'x1 = {equilibrium_x1[0]:.3f}')
plt.axvline(equilibrium_x1[-1], linestyle='--', color='red', label=f'x1 = {equilibrium_x1[-1]:.3f}')
plt.ylim(0, 1.4)
plt.xlim(0, 1)
plt.legend()
plt.xlabel("x1 (Mole Fraction of Component 1)")
plt.ylabel("x * gamma")
plt.show()
```

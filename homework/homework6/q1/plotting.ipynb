```python
import numpy as np
from matplotlib.font_manager import fontManager
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Diodrum W03 Semibold']
plt.rcParams['font.serif'] = sorted(set([f.name for f in fontManager.ttflist if 'serif' in f.name.lower()]))


plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Diodrum W03 Semibold"]
})

A1 = 5450.05
A2 = 5624.58

x1 = np.linspace(0.0001,0.9999,1000)
x2 = 1-x1
gex1 = A1 * x1 * x2
gex2 = A2 * x1 * x2

fig, ax = plt.subplots(dpi=300)

ax.plot(x1, gex1, label='T = 100')
ax.plot(x1, gex2, label='T = 110')
ax.set(xlabel='$x_1$', ylabel=r'$\underline{G}^{ex}$', xlim=(0,1), ylim=0)
ax.grid(zorder=-5, alpha=0.3)
ax.legend();
```

```python
sex = (A2 - A1)/(100-110) * x1 * x2

fig, ax = plt.subplots(dpi=300)

ax.plot(x1, sex, label='T = 105')
ax.set(xlabel='$x_1$', ylabel=r'$\underline{S}^{ex}$', xlim=(0,1))
ax.grid(zorder=-5, alpha=0.3)
ax.legend();
```

```python
Hex = (((A1 + A2)/2) * x1 * x2)  + sex

fig, ax = plt.subplots(dpi=300)

ax.plot(x1, Hex, label='T = 105')
ax.set(xlabel='$x_1$', ylabel=r'$\underline{H}^{ex}$', xlim=(0,1))
ax.grid(zorder=-5, alpha=0.3)
ax.legend();
```

```python
from scipy.constants import R
fig, ax = plt.subplots(dpi=300)


alpha = 1.70001
beta = 2.88792
T1 = 273.15 + 100
gexrt1 = alpha * x1 * beta * x2 / ( alpha*x1 + beta*x2 ) * R * T1
ax.plot(x1, gexrt1, label='T = 100 C')

alpha = 1.44174
beta = 2.58168
T2 = 273.15 + 110
gexrt2 = alpha * x1 * beta * x2 / ( alpha*x1 + beta*x2 ) * R * T2
ax.plot(x1, gexrt2, label='T = 110 C')

ax.set(xlabel='$x_1$', ylabel=r'$\underline{G}^{ex}$', xlim=(0,1), ylim=(0,2000))
ax.grid(zorder=-5, alpha=0.3)
ax.legend();
```

```python
hex = -(((gexrt2 / (T2)) - (gexrt1 / (T1)))/(10)) * (273.15 + 105)**2

fig, ax = plt.subplots(dpi=300)

ax.plot(x1, hex, label='T = 105 C')

ax.set(xlabel='$x_1$', ylabel=r'$\underline{H}^{ex}$', xlim=(0,1), ylim=0)
ax.grid(zorder=-5, alpha=0.3)
ax.legend();
```

```python
sex = (hex - (gexrt1 + gexrt2)/2)/(273.15 + 105)

fig, ax = plt.subplots(dpi=300)

ax.plot(x1, sex, label='T = 105 C')

ax.set(xlabel='$x_1$', ylabel=r'$\underline{S}^{ex}$', xlim=(0,1), ylim=(0,23))
ax.grid(zorder=-5, alpha=0.3)
ax.legend();
```

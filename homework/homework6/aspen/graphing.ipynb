```python
import numpy as np
import matplotlib.pyplot as plt

alpha = np.log(2.17)
beta = np.log(1.98)

def vanlaar_gamma1(x1):
    x2 = 1-x1
    return np.exp(alpha/((1 + (alpha*x1)/(beta*x2))**2))

def vanlaar_gamma2(x1):
    x2 = 1-x1
    return np.exp(beta/((1 + (beta*x2)/(alpha*x1))**2))

fig, ax = plt.subplots(dpi=300)

x1 = np.linspace(0.0001, 0.9999, 10000)
x1g1 = x1 * vanlaar_gamma1(x1)
x1g2 = (1-x1) * vanlaar_gamma2(x1)
plt.plot(x1, x1g1)
plt.plot(x1, x1g2)
ax.set(xlabel='$x_1$', ylabel='$x_1 \gamma_1$, $x_2 \gamma_2$')
```
